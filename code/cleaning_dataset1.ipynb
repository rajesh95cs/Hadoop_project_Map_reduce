```python
#importing packages for cleaning for the Mortality rate dataset
import pandas as pd 
```

```python
import numpy as np
```

```python
#reading the csv file and displaying the first 5 rows
df = pd.read_csv('/home/rajesh/anaconda3/HFAMDB_389_EN.csv')
df.head()
```

```python
#droping the unwanted columns and changing the names of the dataframe
to_drop = ['Unnamed: 1','Unnamed: 2','Unnamed: 3']
df.drop(to_drop,inplace = True, axis = 1)
```

```python
df = df.rename(columns={"Unnamed: 0": "COUNTRY", "Unnamed: 4": "YEAR","Unnamed: 5":"VALUE"})
```

```python
df.head()
```

```python
#filtering out the unwanted 
df = df.sort_values('YEAR',ascending=True)
df = df[df['YEAR'] > '1997']
to_del = df[df['COUNTRY'].isin(['ALB','AND','BIH','GBR','GEO','KAZ','SRB'])].index.tolist()
to_del


```

```python
df = df.drop(index=to_del)
```

```python
#filtering out the necessary data of necessary years
filterd_df = df[df['YEAR']>='1998']
filterd_df
filterd_df.to_csv("Mortality_rate.csv") 
```

```python
#removing the unwanted countries from the dataset
#to match the other dataset
# indexnames stores the index of the values 
# where the countries that are not required for the analysis
# str.contains() checks the record where the value given as the argument 
# is present
# .index returns the index of that particular record
# filterd_df.drop(index=indexnames) drops the records in the dataframe
# that are not required
indexnames = filterd_df[filterd_df['COUNTRY'].str.contains("LTU")].index
filterd_df = filterd_df.drop(index=indexnames)
filterd_df
```

```python
# index stores the order of the records in the dataframe and the 
# index list is added as the first column
index = []
count = 0
for i in range(len(filterd_df)):
    index.append(count)
    count += 1
filterd_df = filterd_df.sort_index()
len(filterd_df)
```

```python
# column is named as index since it was added as the new column
filterd_df['index'] = index
filterd_df =  filterd_df.set_index('index')
filterd_df
```

```python
# dataframe is converted to csv file to check the cleaned data
# format
filterd_df.to_csv('mortality_rate.csv')
```

```python
#csv file is read to a dataframe to be cleaned further
filterd_df = pd.read_csv('mortality_rate.csv')
filterd_df
```

```python
# YEAR column in the dataframe is converted to string type 
filterd_df['YEAR'] = filterd_df['YEAR'].astype('string')
```

```python
# grouping the data to avoid redundancy by taking the mean value
filterd_df = filterd_df.groupby(["COUNTRY","YEAR"])["VALUE"].mean()
filterd_df
```

```python
#cleaned data is exported to a csv file
filterd_df.to_csv('Mortality_cleaned.csv')
```
