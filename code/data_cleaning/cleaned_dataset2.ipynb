```python
# importing packages to clean emigration dataset
import pandas as pd
import numpy as np
```

```python
#read the csv file and store it in a dataframe
df1 = pd.read_csv('migr_emi1ctz.tsv')
df1



```

```python
#droping unwanted columns  
df1 = df1.drop(columns=['citizen','agedef','age','unit','sex'])
df1.head()
```

```python
#droping the unwanted records associated with particular countries
to_del = df1[df1['geo\\time'].isin(['EL','LI'])].index.tolist()
df1 = df1.drop(index=to_del)
```

```python

indexnames = df1[df1['geo\\time'].str.contains("LT")].index
df1 = df1.drop(index=indexnames)
df1
```

```python
#adding index column to the dataframe
df1 = df1.sort_values(by='geo\\time')
index = []
count = 0
for i in range(len(df1)):
    index.append(count)
    count += 1
df1['index'] = index
df1 = df1.set_index('index')
```

```python
#column geo time is renamed as country
df2 = df2.rename(columns={'geo\\time':'COUNTRY'})
df2.head()
df2
```

```python
# A new dictionary df3_dict is created to transpose certain columns
# without affecting other values in the dataset for the purpose of 
# analysis
COUNTRY = df2['COUNTRY'].tolist()
YEAR = df2.columns.tolist()
YEAR.pop(0)
df3_dict = {}


for i in range(0,len(COUNTRY)):
    df3_dict[df2['COUNTRY'][i]] = {}
    for j in range(0,len(YEAR)):
        df3_dict[COUNTRY[i]].update({YEAR[j]:[]})
        
print(df3_dict)

#the dictionary will contain the {'COUNTRY':{'YEAR':['values']}}
#eg sample AUT': {'2017': [], '2016': [], '2015': [], '2014': [], 
#'2013': [], '2012': [], '2011': [], '2010': [], '2009': [], 
#'2008': [], '2007': [], '2006': [], '2005': [], '2004': [], 
#'2003': [], '2002': [], '2001': [], '2000': [], '1999': [],
#'1998': []} the empty list will contain the values with respect 
# to its country and years



    
```
